# Filling Null Values

```python
df.director = df.director.fillna('unknown')
df.cast = df.cast.fillna('unknown')
df.country = df.country.fillna('unknown')
df.date_added = df.date_added.fillna('unknown')
df.rating = df.rating.fillna('unknown')
df.duration = df.duration.fillna('unknown')
```

# Visualization Of The Data

Pie chart of data.
It represent that movies have more value counts than the TV Shows.

```python
plt.pie(df.type.value_counts(),autopct='%1.1f%%',shadow=True)
plt.legend(df.type)
```

Checking "rating" value counts

```python
df.rating.value_counts()
```

Pie Chart of "rating" shows data of rating percentage

```python
plt.pie(df.rating.value_counts(),autopct='%1.1f%%',shadow=True)
plt.legend(df.rating)
```

Barplot of "rating" comparison to the "release_year" 

```python
sns.barplot(x=df.release_year.head(50),y=df.rating)
```

Barplot of "country" 

```python
sns.barplot(x=df.country.value_counts().head(10).values,y=df.country.value_counts().head(10).index)
```

Total counts Analysis via Barplot of "rating" 

```python
sns.barplot(x=df.rating.value_counts().head(10).values, y=df.rating.value_counts().head(10).index)
plt.title('Rating Analysis')
plt.xlabel('Ratings')
plt.ylabel('Counts')
```

Analysis of "Duration" via Barplot 

```python
sns.barplot(y=df.duration.value_counts().head(10).index, x=df.duration.value_counts().head(10).values)
plt.title('Duration Analysis')
plt.xlabel('Count')
plt.ylabel('Duration')
plt.show()
```

Analysis of "rating" via Pie Chart

```python
plt.pie(df[df.country=='India'].rating.value_counts().head(3), labels= df[df.country=='India'].rating.value_counts().head(3).index , autopct='%.1f%%')
plt.title('Rating Distribution in India')
plt.legend()
plt.show()
```

Analysis of "release_year" with the help of Countplot 

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='release_year', data=df)

plt.title('Content Release Over the Years on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()
```

Analysis of "release_year" with help of Kdeplot (Hued = "type")

```python
sns.kdeplot(data=df, x=df.release_year,hue='type',fill= True)
```

Visualization of "df" with help of "Pairplot"

```python
sns.pairplot(df,height=5)
```

Barplot visualization of "title"

```python
sns.barplot(y=df.title.value_counts().head(10).index, x=df.title.value_counts().head(10).values)
plt.title('Title Analysis')
plt.xlabel('Count')
plt.ylabel('Title')
plt.show()
```

Visualization of "type" according to the "release_year" with help of Lineplot

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.type,y=df.release_year,data=df)
```

Total counts of Top 10 directors with the help of "Barplot"

```python
sns.barplot(x=df.director.value_counts().head(10).values,y=df.director.value_counts().head(10).index)
```
